```python
import random
import numpy as np
import matplotlib.pyplot as plt

lista = []
cont = 0
N = 16

for i in range(N):
    u1 = random.random()
    u2 = random.random()
    for j in range(100):
        if u2 <= np.sqrt(1 - u1**2):
            cont = cont + 1
            plt.scatter(u1, u2, color="red")
        else:
            plt.scatter(u1, u2, color="blue")

x = np.linspace(0, 1, 100)
plt.plot(x, np.sqrt(1 - x**2), label="Quarter circle")
plt.xlabel("u1")
plt.ylabel("u2")
plt.title("Monte Carlo Simulation for π estimation")
plt.legend()
plt.show()

p = cont / (N * 100)
pi_estimate = 4 * p
lista.append(pi_estimate)

print(f"Estimación de π: {pi_estimate}")
```
