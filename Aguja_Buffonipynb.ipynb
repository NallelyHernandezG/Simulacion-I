```python
import numpy as np
import matplotlib.pyplot as plt

def buffon_needle_simulation(num_trials, needle_length, line_distance):
    crossings = 0

    pi_estimates = []

    for _ in range(num_trials):
        # Randomly drop the needle
        center = np.random.uniform(0, line_distance / 2)
        angle = np.random.uniform(0, np.pi / 2)

        # Check if the needle crosses a line
        if center <= (needle_length / 2) * np.sin(angle):
            crossings += 1

        # Estimate pi
        if crossings > 0:
            pi_estimate = (2 * needle_length * num_trials) / (line_distance * crossings)
        else:
            pi_estimate = np.nan  # Handle case where crossings is 0

        pi_estimates.append(pi_estimate)

    return pi_estimates

# Parameters
num_trials = 10000
needle_length = 1.0
line_distance = 1.0

# Run the simulation
pi_estimates = buffon_needle_simulation(num_trials, needle_length, line_distance)

# Plot the simulation results
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, label='Estimated π')
plt.axhline(y=np.pi, color='r', linestyle='-', label='Actual π')
plt.xlabel('Number of Trials')
plt.ylabel('Estimated Value of π')
plt.title('Buffon\'s Needle Simulation')
plt.legend()
plt.grid(True)
plt.show()
```
