```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definición de la función a integrar
def f(x):
    return np.arctan(x)

# Parámetros de la simulación
n_samples = 10000  # Número de muestras
a, b = 0, 1  # Límites de integración

# Generación de muestras aleatorias
x_samples = np.random.uniform(a, b, n_samples)
y_samples = f(x_samples)

# Cálculo de la integral mediante el método de media muestral
integral_estimate = (b - a) * np.mean(y_samples)

# Intervalo de confianza del 95%
std_error = np.std(y_samples) / np.sqrt(n_samples)
z = norm.ppf(0.975)  # Valor crítico para un intervalo de confianza del 95%
confidence_interval = (
    integral_estimate - z * std_error,
    integral_estimate + z * std_error,
)

# Resultados
print(f"Estimación de la integral: {integral_estimate:.5f}")
print(f"Intervalo de confianza del 95%: ({confidence_interval[0]:.5f}, {confidence_interval[1]:.5f})")

# Gráfica de la función y puntos de muestra
x = np.linspace(a, b, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='arctan(x)', color='blue')
plt.scatter(x_samples, y_samples, color='red', s=1, alpha=0.5, label='Muestras')
plt.title('Integración de arctan(x) mediante Monte Carlo')
plt.xlabel('x')
plt.ylabel('arctan(x)')
plt.legend()
plt.grid(True)
plt.show()

# Histograma de los valores de las muestras
plt.figure(figsize=(10, 6))
plt.hist(y_samples, bins=50, density=True, alpha=0.75, color='green')
plt.title('Histograma de las evaluaciones de arctan(x)')
plt.xlabel('arctan(x)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
