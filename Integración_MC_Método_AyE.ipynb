```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de la funci√≥n a integrar
def f(x):
    return np.arctan(x)

# Par√°metros de la simulaci√≥n
n_samples = 10000  # N√∫mero de muestras
a, b = 0, 1  # L√≠mites de integraci√≥n
y_min, y_max = 0, np.arctan(1)  # Valores m√≠nimos y m√°ximos de y en el rango [0, 1]

# Generaci√≥n de muestras aleatorias
x_samples = np.random.uniform(a, b, n_samples)
y_samples = np.random.uniform(y_min, y_max, n_samples)

# Evaluaci√≥n de los aciertos (puntos bajo la curva)
under_curve = y_samples <= f(x_samples)
hits = np.sum(under_curve)

# C√°lculo de la integral mediante el m√©todo de acierto y error
area_rect = (b - a) * (y_max - y_min)
integral_estimate = area_rect * (hits / n_samples)

# Resultados
print(f"Estimaci√≥n de la integral: {integral_estimate:.5f}")

# Gr√°fica de la funci√≥n y puntos de muestra
x = np.linspace(a, b, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='arctan(x)', color='blue')
plt.scatter(x_samples, y_samples, color='red', s=1, alpha=0.5, label='Puntos bajo la curva' if under_curve.all() else 'Puntos fuera de la curva')
plt.title('Integraci√≥n de arctan(x) mediante Monte Carlo (Acierto y Error)')
plt.xlabel('x')
plt.ylabel('arctan(x)')
plt.legend()
plt.grid(True)
plt.show()

# Gr√°fico de los puntos bajo y sobre la curva
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='arctan(x)', color='blue')
plt.scatter(x_samples[under_curve], y_samples[under_curve], color='green', s=1, alpha=0.5, label='Puntos bajo la curva')
plt.scatter(x_samples[~under_curve], y_samples[~under_curve], color='red', s=1, alpha=0.5, label='Puntos fuera de la curva')
plt.title('Puntos bajo y fuera de la curva de arctan(x)')
plt.xlabel('x')
plt.ylabel('arctan(x)')
plt.legend()
plt.grid(True)
plt.show()
```
